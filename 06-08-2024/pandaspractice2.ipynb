Find the most recent product entry based on the 'last_modified_datetime' column.

```python
food['last_modified_datetime']=pd.to_datetime(food['last_modified_datetime'],errors='coerce')
foodd = food
foodd=foodd.dropna(subset=['last_modified_datetime'])
rec = foodd['last_modified_datetime'].idxmax()
foodd.loc[rec,['code','product_name','last_modified_datetime']]
```
